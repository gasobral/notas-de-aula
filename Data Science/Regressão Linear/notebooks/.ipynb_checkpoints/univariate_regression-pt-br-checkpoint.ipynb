# Regressão Linear
A *regressão linear* é um modelo bastante utilizado pois ele é intuitivo
e sua interpretação não é complicada. Por exemplo, temos a intuição de
que o gasto de uma pessoa tem uma relação com a sua renda. Para
quantificarmos essa relação, podemos utilizar uma regressão a fim de
expressar o gasto em função da renda. Podemos escrever isso como

$$ gasto = renda \times \beta$$, onde $\beta$ é um valor real.

Da forma como escrevemos a conta, estamos assumindo que exista uma
proporção, expresso pelo $\beta$, entre gasto e renda, em outras
palavras, que existe uma relação linear entre renda e gasto. Esse fato é
bastante importante na regressão linear.

Agora, para estabelecermos um modelo a partir desta conta, vamos
adicionar um valor real $\alpha$, nesse contexto ele representa o gasto
quando a renda é zero, e um valor real $e$ para representar o erro do
modelo. Desta forma, temos o seguinte:

$$ gasto = \alpha + renda \times \beta + e$$

Agora o nosso modelo está pronto. Note que podemos usá-lo para prever o
gasto a partir da renda. Desde que exista uma forte relação linear
(expresso pelo coeficiente de correlação) entre as variáveis e o erro
não seja grande. Algumas observações sobre o modelo:

- o gasto é a variável alvo (o que desejo prever);
- a renda é a variável preditora (variável usada para prever valores);
- o $\alpha$ e o $\beta$ são os parâmetros do modelo.

Você também pode encontrar a nomenclatura variável explicadora para
variável preditora. Mais adiante falaremos sobre os parâmetros do
modelo. No momento basta saber que, quando um modelo é treinado, são
determinados parâmetros que minimizem o erro $e$. Após vermos esse
exemplo, não é tão assustador definir a regressão linear como:

$$ y = \alpha + \beta x + e$$, onde $y$ é variável alvo e $x$ a variável
preditora. Adiante apresentarei uma situação problema para abordamos a
regressão linear na prática. Por enquanto estamos lidando apenas com uma
variável, mas abordaremos uma generalização desse modelo. Usarei como referência os livros
[^1], [^2] e [^3] para apresentar este modelo de forma exploratória.

[^1]: BUSSAB, W. O.; MORETTIN, P. A. Estatística Básica. Atual Editora, São Paulo, 1988

[^2]: GÉRON, Aurélien. Hands-On Machine Learning with Scikit-Lean, Keras and TensorFlow. O' Reilly Media Inc. 2021

[^3]: MORETTIN, Pedro Alberto; SINGER, Júlio da Motta. Estatística e
ciência de dados. Rio de Janeiro, LTC, 2022

## Situação-problema
Uma empresa da área da saúde deseja monitorar a função respiratória de
funcionários expostos à poeira. Ela forneceu uma planília com os
seguintes dados:

| Nome da coluna | Descrição                           |
| :---           | :---                                |
| PEFR           | taxa do volume de ar expelido dos   |
|                | pulmões em uma expiração rápida,    |
|                | esse é um indicador confiável da    |
|                | ventilação e também do fluxo de ar. |
| Exposure       | tempo de exposição em anos à poeira.|

Deseja-se estimar o PEFR dado o tempo de exposição (Exposure).

A primeira tarefa a ser feita é entendermos os dados que estamos
lidando. O Exposure é bastante direto e não precisa de muitas
observações. O PEFR, aparentemente, é uma medida do universo dos números
reais. Uma vez identificado o domínio dos dados, vamos examiná-los
usando a análise exploratória (hora de usarmos os conhecimentos em
estatística).

```python
import pandas as pd
```

```python
file_path = "../data/LungDisease.csv"
lung_disease = pd.read_csv(file_path)
```

```python
lung_disease.info()
```

csv.

- tipo de dados
- valores ausentes

No problema em que estamos trabalhando, o PEFR foi carregado como

valores ausentes.

```python
## uma pequena amostra dos dados para termos uma noção
## mais concreta deles
lung_disease.head(5)
```

Agora vamos analisar a distribuição dos dados.

```python
lung_disease.describe().T
```

A partir dessas medidas descritivas, o PEFR aparenta ter uma
distribuição similar a da distribuição normal. Pois o primeiro quartil e
o terceiro quartil estão aproximadamente a um desvio padrão da
mediana. No entanto, o mínimo e o máximo não estão (ambos) a dois

está a um desvio padrão da média. Essas observações nos ajudam a
compreender os dados que estamos lidando e, no futuro, a definirmos
estratégias de normalização dos dados - que poder ser bem útil para um
modelo -, por exemplo. Agora, para essa discussão não ficar muito

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 8))

ax[0].hist(lung_disease['PEFR'])
ax[0].set_xlabel('Taxa do pico do fluxo respiratório')
ax[0].set_ylabel('Frequência')
ax[0].set_title('Histograma do PEFR')

ax[1].hist(lung_disease['Exposure'])
ax[1].set_xlabel('Exposição em anos')
ax[1].set_ylabel('Frequência')
ax[1].set_title('Histograma da Exposição em anos')

plt.show()
```

Olhando para os histogramas fica bem mais simples de visualizarmos as

Agora, como queremos prever o PEFR a partir dos anos de exposição, então
estamos interessado na relação entre essas duas variáveis. Como se
tratam de variáveis numéricas, podemos usar um gráfico de dispersão para
mostrarmos visualmente a relação entre PEFR e o tempo de exposição.

```python
fig, ax = plt.subplots(1, 1, figsize=(14, 8))

ax.scatter(x=lung_disease['Exposure'],
           y=lung_disease['PEFR'])
ax.set_xlabel('Exposição em anos')
ax.set_ylabel('PEFR')
ax.set_title('Gráfico de dispersão da Exposição x PEFR')

plt.show()
```

No gráfico acima, devemos procurar por algum padrão que possa descrever
a relação entre as variáveis em questão. Veja que, de modo geral, um
aumento no tempo de exposição reduz o PEFR. Esse tipo de situação tende
a formar uma reta descente, indicando que possa haver uma relação entre
PEFR e o tempo de exposição. Porém, no tempo de exposição igual a 20, há
linha paralela ao eixo y (PEFR). Isso pode ser ruídos, erros de
digitação ou de alguma outra variável que afeta PEFR e o tempo de
exposição. Em todos os casos, os dados que temos não nos permitem
aprofundarmos nessa análise. Podemos, por fim, usarmos o coeficiente de
correlação para expressar a correlação linear entre PEFR e o tempo de
exposição.

```python
lung_disease['PEFR'].corr(lung_disease['Exposure'])
```

Não é uma correlação negativa muito intensa, mas nos permite afirmarmos
que o tempo de exposição tende a diminuir o PEFR. Portanto, podemos usar
o tempo de exposição para prevermos o valor do PEFR.

Nessa primeira etapa, fizemos o essencial:

- entendemos o problema;
- identificamos as estruturas dos dados;
- fizemos uma análise exploratória dos dados;
- determinamos quais variáveis podemos usar para a predição.

Agora estamos prontos para criarmos o modelo.

## Univariate Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
X = lung_disease.drop(columns=['PEFR'])
y = lung_disease['PEFR']
```

```python
linear_model = LinearRegression()
linear_model.fit(X, y)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(14, 8))

ax.set_xlim(0, 23)
ax.set_ylim(0, 610)
# ax.set_ylim(295, 450)

ax.set_xlabel('Exposure')
ax.set_ylabel('PEFR')

## obtém os valores extremos da equação de regressão para desenhar os
## extremos da reta
ax.plot((0, 23),
        linear_model.predict(pd.DataFrame({'Exposure': [0, 23]})))

## posiciona o intercepto (coeficiente b_0) na reta
ax.text(0.4,
        linear_model.intercept_,
        r'$b_0$',
        size='larger')

plt.show()
```

```python
print(f'Equation: Y = {linear_model.intercept_:.3f} + '
      f'{linear_model.coef_[0]:.3f} X')
```

```python
predicted_value = linear_model.predict(X)
error = y - predicted_value

for x, real, p, erro in zip(lung_disease.Exposure,
                            lung_disease.PEFR,
                            predicted_value,
                            error):
    print(f"X: {x}  Real value: {real}  Predicted value: {p}  Error: {erro}")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(14, 8))

ax.scatter(x=lung_disease['Exposure'],
           y=lung_disease['PEFR'])

ax.plot(lung_disease.Exposure, predicted_value)

for x, y_real, y_predicted in zip(lung_disease.Exposure,
                                  lung_disease.PEFR,
                                  predicted_value):
    ax.plot((x,x), (y_real, y_predicted), '--', color='C1')

ax.set_title('Predicted value x Real value')
ax.set_xlabel('Exposure in years')
ax.set_ylabel('PEFR')

plt.show()
```
